### 심파이를 사용한 함수 미분

```python
# 예측 모형의 성능 - 함수의 계수(=모수,가중치)에 달려있다.
#   따라서 모수를 결정하여 성능을 측정하는 과정은 다변수 함수를 계산하는 것과 같음. = 성능함수
#   성능함수의 값은 클수록 좋음
#   반대로 모수를 입력받아 오차 혹은 오류의 크기를 출력하는 함수를 가장작게 할 수 도 있음
#   = 손실함수 or 비용함수
#   이런 최적화의 대상이 되는 함수를 통틀어 목적함수라고 함.

# 최적화 - 목적함수의 값을 가장 크게 혹은 작게 할 수 있는 입력값을 구하는것
#   최적화를 위해 입력값에따른 출력값의 변화를 정량적으로 알아야함 = 미분

# 기울기 - 수치적 최적화를 위해 필요함.(가장 적은 횟수로 x값을 시도하여 최적의 값을 찾아냄)
#          어떤 x1을 시도 했으면 다음 x2값으로 뭐가 좋을지 선정하기 위함.
#          기울기는 이 시도의 방향을 알기 위해 사용됨. 더 큰 수 혹은 작은 수
# Δy/Δx = (f(x2)-f(x))/(x2-x) = (f(x+Δx)-f(x))/Δx
# slope(기울기)=limΔx→0 (f(x+Δx)-f(x))/Δx

# + 방정식 계산 - 0=ax^n+bx^(n-1)+....+wx+y > np.roots([a,b,...,w,y])으로 구현되어있음
```

```python
# 수치미분 derivative()로 대략적인 기울기를 계산 가능
# 인수: 함수f, 위치x, 이동거리dx
# slope ~ (f(x+dx/2)-f(x-dx/2))/dx = 수치미분
# 이렇게 구해진 기울기는 대략적. 정확한 값은 구할 수 없음 dx는 작을 수록 좋지만 너무 작으면 오류남
```

```python
# 미분 - 어떤 함수로부터 그 함수 기울기를 출력하는 새로운 함수를 만들어내는 작업
# 도함수 - 미분으로 만들어진 새로운 함수
#       f′=d/dx *(f)= df/dx = d/dx *y = dy/dx

# 미분가능
# 앞에서 봤던 ReLU함수는 x=0에서 미분 불가능하다. 추가로 내용을 정리해보면
# 부드러운 연속, 좌미분계수=우미분계수 이면 미분가능하다.

# 미분공식
# 1. 기본미분공식
#    상수 - 미분시 0 | 거듭제곱 d/dx *x^n = nx^(n-1) | 로그 d/dx (logx)=1/x | 지수 d/dx e^x =e^x
# 2. 선형조합법칙
#    그냥 분배법칙이 된다고 생각하면됨
# 3. 곱셈법칙
#    d/dx*(f*g)=f*dg/dx+df/dx*g 
# 4. 연쇄법칙
#    f(x)=h(g(x)) ->  df/dx=(dh/dg)*(dg/dx)
# 그냥 햇갈릴거 같은거
# d/dx logf(x)= f′(x)/f(x) | d/dx e^f(x) = f′(x)*e^f(x)

# 2차 도함수 - 도함수를 한 번 더 미분한 함수 = d^2/dx^2 
# 당연히 도함수의 기울기를 나타냄
# 왜인지는 모르겠지만 f′′>0 이면 볼록, f′′<0 이면 오목하다 함.
```

```python
# 편미분 - 다변수 함수인 경우에 하나에 변수에 대한 기울기를 구하는 방법
#   -> 편미분의 결과로 하나으 ㅣ함수에 대해 여러 개의 도함수가 나올 수 있음
#   표기: f_x(x,y) = ∂f/∂x | f_y(x,y) = ∂f/∂y
#   방법: 다른 독립변수를 상수로 취급하고 미분하면 됨

# 다변수 함수의 연쇄법칙 - 연쇄법칙 + 편미분

# 2차 편미분 - 편미분으로 만든 2차 도함수
#   f_xx(x,y)= ∂^2*f/∂x^2 | f_yy(x,y)=∂^2*f/∂y^2 |
#   f_xy(x,y)=∂^2*f/∂y∂x (첫번째 x에대해 두번째 y에대해)
#   f_yx(x,y)=∂^2*f/∂x∂y  - 사실 f_xy = f_yx 임 순서 상관없음

# 테일러 전개 - 함수모양의 근사화 (x0)
# 특정 조건을 만족하는 무한번 미분 가능한 함수 f(x)에 대해
# f(x)= ∑k=0~∞ f^(k)(0)/k! *x^k 가 성립. 여기서 f^(k)는 k번 미분 했다는 뜻.
# e^x의 테일러 전개
#   e^x = a_0+a_1x+a_2x^2+...로 표현 시 x=0을 넣으면 a_0=1
#   미분하면 e^x=(1)a_1+(1*2)a_2x+... 에 x=0을 넣으면 a_1=1/1
#   다시 미분하면 e^x=(1*2)a_2+(1*2*3)a_3x^+... x=0 이면 a_2=1/2!
#   즉 e^x=1+x+x^2/2!+x^3/3!+... 의 형태
#   일반식으로 전개시 a_i = f^(i)(0)/i! 이 됨. (f^(i)는 i번 미분했다는 뜻)
#   즉 일반식은 f(x)=∑k=0~∞ f^(k)(0)/k! *x^k
# 코딩으로는 f(x)≈f(x0)+df(x0)/dx*(x−x0)이 됨. x0는 x의 위치.
```

```python
# 심파이 - 심볼릭 연산(미적분)을 지원하는 패키지
import sympy; import numpy as np
x=sympy.symbols('x') # 변수를 심볼로 지정해줘야 연산이 가능함 ㅅㅂ
f=x*sympy.exp(x)
sympy.diff(f) #미분
sympy.simplify(sympy.diff(f)) # 미분+소인수분해
x,y=sympy.symbols('x y')
f=x**2+4*x*y+4*y**2
sympy.diff(f,x) # x에 대해 편미분
# 상수 심볼을 포함하는 함수를 미분하는 경우 편미분처럼 심볼과 입력 변수를 지정해줘야함
x,mu,sigma=sympy.symbols('x mu sigma')
f=sympy.exp((x-mu)**2/sigma**2); f
sympy.diff(f,x)
sympy.diff(f,x,x) # 2차 도함수
```

### 적분

```python
# 적분 - 미분과 반대되는 개념. 부정적분 과 정적분 으로 나뉨

# 부정적분 - 정확하게 미분과 반대개념. 즉 반-미분임.
# 도함수를 f(x)라고하면 미분 전 함수를 F(x) 혹은 ∫f(x)dx 라고 씀
# F(x)=∫f(x)dx + C 

# 편미분의 부정적분
# f(x,y)가 F(x,y)를 x로 편미분한 함수이면 = ∂F(x,y)/∂x = f(x,y)
# F(x,y)=∫f(x,y)dx +C(y)이다. 주의할 점은 상수항C가 y의 함수일 수 있다는것.(상수처리했기 때문)

# 다차 도함수와 다중적분
# 만약 f(x,y)가 F_3(x,y)를 x -> y 순으로 편미분한 이차 도함수라면
# f(x,y)를 y로 한번, x로 한번 적분하면 F_3(x,y)가 나온다. = ∬f(x,y)dydx
# 근데 이러면 변수일수도 있는 상수항 처리는 어떻게 하노?

# 심파이를 이용한 부정적분 - integrate()
import sympy as sym; import scipy as sci
x=sym.symbols('x')
f=x*sym.exp(x)+sym.exp(x); f
sym.integrate(f)
x,y=sym.symbols('x y')
f=2*x + y; f
sym.integrate(f,x)
```

```python
# 정적분 
# - 독립변수 x가 구간[a,b] 사이일 때, 그 구간에서 함수 f(x)의 값과 수평선이 이루는 면적을 구함
#   ∫b~a f(x)dx =F(b)=F(a)  - 미적분학의 기본 정리
#   부정적분을 한뒤 기본정리를 사용하거나, 면적을 잘게쪼개 면적을 근사하여 구하는 수치적분으로 나뉨
# + zip은 동일한 원소갯수,동일한 크기로 이루어진 자료형을 튜플로 묶은 리스트화시킴
from matplotlib.patches import Polygon; import matplotlib.pyplot as plt
def f(x):
    return x**3-3*x**2+x+6
a,b=0,2
x=np.linspace(a-0.5,b+0.5,50); y=f(x)
ax=plt.subplot(111) 
plt.title('ex)definite integral'); plt.plot(x,y,'r',lw=2); plt.ylim(bottom=0)
ix=np.linspace(a,b); iy=f(ix)
verts = [(a,0)]+list(zip(ix,iy))+[(b,0)]
poly=Polygon(verts,facecolor='0.9',edgecolor='0.5')
ax.add_patch(poly)
plt.text(0.5*(a+b),0.2*(f(a)+f(b)),r'$\int_a^b f(x)dx$',horizontalalignment='center',fontsize=20)
plt.figtext(0.9,0.05,'$x$'); plt.figtext(0.1,0.9,'$y$')
ax.spines['right'].set_visible(False); ax.spines['top'].set_visible(False)
ax.xaxis.set_ticks_position('bottom')
ax.set_xticks([a,b]); ax.set_xticklabels(['$a$','$b$'])
ax.set_yticks([]); ax.set_xlim(-2,4); ax.set_ylim(0,8)
plt.show()
x,y=sym.symbols('x y')
f=x**3-3*x**2+x+6
F=sym.integrate(f); F
(F.subs(x,2)-F.subs(x,0)).evalf() # subs, evalf 심볼릭 변수에 실제 숫자를 넣어 계산하기 위한 매서드

# 수치적분 - 함수를 아주 작은 구간으로 나누어 실제 면적을 계산
# quad 명령 사용
from scipy.integrate import quad
def f(x):
    return x**3-3*x**2+x+6
quad(f,0,2) # 결과 값의 두번째 숫자는 오차 상한값을 의미
```

```python
# 다변수 정적분 - 다양한 방법으로 정의 가능
# 만약 두 변수 모두 적분하는 것은 2차원 평면에서 주어진 사각형 아래의 부피를 구하는 것과 같음
# ex) ∫y=0~6∫x=2~8 x^2−10x+y+50 dxdy
fig=plt.figure()
ax=fig.add_subplot(111,projection='3d')
_x=np.arange(12)/2+2
_y=np.arange(12)/2
X,Y=np.meshgrid(_x,_y)
x,y=X.ravel(),Y.ravel()
z=x*x-10*x+y+50
z0=np.zeros_like(z)
ax.bar3d(x,y,z0,0.48,0.48,z)
ax.set_xlim(0,10); ax.set_ylim(-2,10); ax.set_zlim(0,50)
ax.set_xlabel('x'); ax.set_ylabel('y')
plt.title('f(x,y)'); plt.show()
# 수치이중적분을 하려면 scipy.integrate의 dblquad()를 사용
# dblquad(func,a,b,gfun,hfun) - a,b x의 하한,상한 | gfun,hfun y의 하한과 상한 - x의 함수여야함
def f(y, x):
    return np.exp(-x * y) / x**2
from scipy.integrate import dblquad
dblquad(f,1,np.inf,lambda x:0,lambda x:np.inf)

# 다차원 함수의 단일 정적분 - 변수중 하나만 변수로보고 다른 변수는 상수로 취급
```

### 행렬의 미분

```python
# 행렬의 미분 - 다변수 함수는 입력변수가 벡터인 경우로 볼 수 있음
# f(np.array([x1,x2]))=f=(x)=f(x1,x2)  vector -> scalar
# 이러면 행렬을 입력으로 갖는 함수도 당연히 있음 matrix -> scalar

# 반대로 벡터나 행렬을 출력하는 함수는 여러개의 함수를 합쳐놓은것.
# f(x)=np.array([f1(x),f2(x)]) scalar->vector
# f(x)=np.array([[f11(x),f12(x)],[f21(x),f22(x)]]) scalar -> matrix

# 벡터나 행렬을 입력받아 벡터나 행렬 출력도 가능함
# f(x)=np.array([f1(x1,x2),f2(x1,x2)]) vector -> vector
# f(x)=np.array([[f11(x1,x2),f12(x1,x2)],[f21(x1,x2),f22(x1,x2)]]) vector -> matrix

# 이렇게 입출력을 행렬로 갖는 함수를 미분하는 것을 행렬미분이라 함. - 정확히는 편미분임
```

```python
# 스칼라를 벡터로 미분하는 경우 - 결과값은 열벡터로 표시 = 그레디언트 벡터 ∇f
# ∇f = ∂f/∂x = [∂f/∂x1 ∂f/∂x2 ... ∂f/∂xn].T (열벡터)
# 2차원의 경우 그레디언트 벡터는 컨투어플롯으로 나타냄. 
# 컨투어 플롯위에 그레디언트 벡터를 화살표로 나타낸 것을 퀴버플롯이라 함.
# 만약 어떤위치 x,y에서 x증가시 f커지면 도함수∂f/∂x는 양수,y증가시 f작아지면 ∂f∂y는 음수
# 그레디언트 벡터의 방향은 편미분 성분의 부호에 의해 결정
def f(x, y):
    return 2 * x**2 + 6 * x * y + 7 * y**2 - 26 * x - 54 * y + 107
xx=np.linspace(1,16,100)
yy=np.linspace(-3,6,90)
X,Y=np.meshgrid(xx,yy)
Z=f(X,Y)
def gx(x,y): return 4 * x + 6 * y - 26
def gy(x,y): return 6 * x + 14 * y - 54
xx2=np.linspace(1,16,15)
yy2=np.linspace(-3,6,9)
X2,Y2=np.meshgrid(xx2,yy2)
GX=gx(X2,Y2)
GY=gy(X2,Y2)
plt.figure(figsize=(10,5))
plt.contour(X,Y,Z,levels=np.logspace(0,3,10))
plt.quiver(X2,Y2,GX,GY,color='blue',scale=400,minshaft=2)
plt.xlabel('x'); plt.ylabel('y'); plt.title('quiver plot')
plt.grid(True); plt.show()
# 퀴버플롯에서 그레디언트 벡터의 특징
#   그레디언트 벡터의 크기 = 기울기. 즉 크기가 클수록 함수 곡면의 기울기가 커짐
#   ∇f의 방향은 함수 곡면의 기울기가 가장 큰 방향. 즉 단위길이당 함수값이 가장크게 증가하는 방향임
#   ∇f의 방향은 등고선의 방향과 직교함
```

```python
# 행렬미분법칙
# 법칙 1. 선형모형
# f(x) = w.tx 의 선형모형을 미분하면 그레디언트 벡터는 가중치 벡터이다.
# ∇f = ∂w.tx/∂x = ∂x.tw/∂x = w

# 법칙 2. 이차 형식
# f(x) =  x.tAx의 이차형식을 미분하면 행렬과 벡터의 곱으로 나타난다.
# ∇f= ∂x.tAx/∂x = (A+A.t)x
# 성분을 나열 후 편미분하면 각 편미분요소는 N행N열만 남게되고
# ∑i=1~N(a_1i*x_i) + ∑i=1~N(a_i1*x_i), ... , ∑i=1~N(a_Ni*x_i) + ∑i=1~N(a_iN*x_i)이 됨
# 앞쪽은 곧 Ax이고 뒤는 곧 A.Tx이므로 Ax+A.Tx=(A+A.T)x

# 쉽게보면 ax-> a , w.Tx -> w | ax^2 -> 2ax, x.tAx -> (A+A.T)x

# 벡터를 스칼라로 미분하는 경우(입력:스칼라, 출력:벡터)
# f(x) = [f1 f2 ... fM].T 이면 ∂f/∂x = [∂f1/∂x ∂f2/∂x ... ∂fM/∂x]가 됨. (행벡터)

# 벡터를 벡터로 미분하는 경우(입/출력:벡터)
# ∂f/∂x = [∂f1/∂x ∂f2/∂x ... ∂fn/∂x] = [∂f1/∂x ∂f2/∂x ... ∂fn/∂x].T
# = [∂f1/∂x1 ∂f2/∂x1 ... ∂fn/∂x1
#    ∂f1/∂x2 ∂f2/∂x2 ... ∂fn/∂x2
#            쭉 내려가면
#    ∂f1/∂xm ∂f2/∂xm ... ∂fn/∂xm]
# 미분 당하는 벡터의 원소가 여러개(i=1~N)이고 미분 하는 벡터의 원소도 여러개(j=1~M)이므로
# 도함수는 행렬이 됨.

# 법칙 3. 행렬과 벡터의 곱의 미분
# f(x)=Ax를 벡터 x로 미분시 ∇f(x)=A.T
# 자코비안 행렬: 입출력이 모두 벡터인 경우 행렬형태의 도함수.
#   벡터함수를 벡터변수로 미분해서 생기는 행렬의 전치행렬.
# 헤시안 행렬: 그레디언트 벡터의 자코비안 전치행렬 Hf(x)=H=J(∇f(x)).T
# H_ij=∂^2*f/∂x_i*∂x_j. 함수가 미분가능하면 대칭행렬이 됨.

# 스칼라를 행렬로 미분(출력:스칼라, 입력:행렬)
# ∂f/∂X = [∂f/∂x_1,1 ... ∂f/∂x_1,n
#          ∂f/∂x_2,1 ... ∂f/∂x_2,n
#             .               .    
#             .               .
#          ∂f/∂x_m,1 ... ∂f/∂x_m,n]

# 법칙 4. 행렬 곱의 대각성분
# f(x)=tr(WX) (W∈RN^2,X∈RN^2)
# ∂f/∂X = ∂tr(WX)/∂X = W.T

# 법칙 5. 행렬식의 로그
# f(X)=log|X| > ∂f/∂X = ∂log|X|/∂X = X^(-1).T
```

### 편분법

```python
# 범함수 - 함수를 입력으로 받아 실수를 출력
#   기댓값, 엔트로피 계산시 함수를 입력받아 실수를 출력함.
#   대문자로 표기. F[y(x)]
#   보통 정적분으로 계산하는 경우가 많음

# 범함수의 테일러 전개
#   일반적인 테일러 전개 = f(x+ϵ)≈f(x)+df/dx*ϵ (ϵ는 아주 작은 실수)
# F(y1+ϵ1,y2+ϵ2,...,yn+ϵn)≈F(y1,y2,...,yn)+∂F/∂y1*ϵ1+∂F/∂y2*ϵ2+...+∂F/∂yn*ϵn
#                          =F(y1,y2,...,yn)+∑i=1~N ∂F/∂yi*ei
# 여기서 yi=y(xi), ϵi=ϵη(xi) 로 가정.
# 그럼 위 식 = F(y(x1),y(x2),…,y(xN)) + ϵ∑i=1~N ∂F/∂yi*η(xi)
# 여기서 {y(x1),⋯,y(xN)} = y(x)로 하며, N을 무한대로 늘리면 범함수가 됨
# 즉, F[y(x)+ϵη(x)]≈F[y(x)]+ϵ∫ δF/δy(x)*η(x) dx

# 범함수의 도함수
# 범함수 테일러 전개에서 ϵ에 대한 변화는 (F[y(x)+ϵη(x)]−F[y(x)])/ϵ = ∫ δF/δy(x)*η(x) dx
# δF/δy(x)=0 이기에 δF/δy(x)를 범함수의 도함수라 함

# 적분형 범함수의 도함수
#   대부분의 범함수는 x에 대한 적분으로 정의됨 
#   적분기호안의 연산은 y(x)와 x를 입력으로 받는 함수 G(y,x)임
#   F[y(x)]=∫G(y,x)dx -> δF/δy=∂G/∂y
```

```python
# 오일러 라그랑주 공식
# x,y(x)이외에 추가로 y(x)의 도함수인 y′(x)를 입력으로 받는 함수 G(y(x),y′(x),x)가 있을 수 있음
# 이때 범함수 F의 도함수는 δF/δy = ∂G/∂y−d/dx*(∂G/∂y′)가 됨

# 최적제어
# 최적화 = 함수가 있을 때 값이 max or min이 되도록 하는 x의 값을 찾아내는 작업
# 최적 제어 = 범함수의 최적화.(F(y(x))가 있을 때 y(x)의 값 찾아내는 작업)
# 최적화의 필요조건 - 최적값 x*를 입력하면 도함수값이 0이 되야함. df/dx*(x*)=0
# 최적 제어의 필요조건 - 최적값 y∗(x)입력시 범함수의 도함수 값이 0
# δf/δy*[y∗(x)]=0  
```
